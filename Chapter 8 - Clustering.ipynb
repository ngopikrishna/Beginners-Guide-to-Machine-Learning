<h1>k-Means Clustering</h1>

```python
import pandas as pd
from sklearn.cluster import KMeans

# Load the datasets
axis_x = 1 # Age
axis_y = 4 # Fare


dataframe = pd.read_csv("titanic.csv")
dataframe = dataframe.drop(columns=['Name', 'Ticket', 'Cabin', 'PassengerId', 'Pclass', 'Sex','Embarked'])
X = dataframe.values
kmeans = KMeans(n_clusters=3) 
kmeans.fit(X)
y_kmeans = kmeans.labels_
print(y_kmeans)


import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()

cluster_markings = ['+', '*', '.']
cluster_point_colours = ['black', 'blue', 'green']
for mnx in range(len(cluster_markings)):
    cluster_points = (y_kmeans==mnx)
    x_values = X[cluster_points,axis_x]
    y_values = X[cluster_points,axis_y]
    plt.scatter(x_values, y_values, c=cluster_point_colours[mnx], marker = cluster_markings[mnx])

plt.scatter(kmeans.cluster_centers_[:, axis_x], kmeans.cluster_centers_[:, axis_y], s=30, c='red', marker='s')
plt.show()
```

<h1> Hierarchical Clustering </h1>

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch


# Load the datasets
axis_x = 1 # Age
axis_y = 4 # Fare


dataframe = pd.read_csv("titanic.csv")
dataframe = dataframe.drop(columns=['Name', 'Ticket', 'Cabin', 'PassengerId', 'Pclass', 'Sex','Embarked'])
X = dataframe.values


dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
model = AgglomerativeClustering( affinity='euclidean', linkage='ward')
model.fit(X)
labels = model.labels_
print(labels)
```

```python
# There is no direct function to plt scree charts in Python for Hierarchichal clustering. 
# A work-around is to create multiple k-means clustering models with k raning from 1 to n
# The kmeans_model.inertia_ parameter provides wssr for that model.
# Then we can plot the scree chart with this data

wcss = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(10, 8))
plt.plot(range(1, 20), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()

cluster_markings = ['+', '*', '.']
cluster_point_colours = ['black', 'blue', 'green']
for mnx in range(model.n_clusters):
    cluster_points = (labels==mnx)
    x_values = X[cluster_points,axis_x]
    y_values = X[cluster_points,axis_y]
    plt.scatter(x_values, y_values, c=cluster_point_colours[mnx], marker = cluster_markings[mnx])

plt.show()
```

<h1>DBSCAN</h1>

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN



dataframe = pd.read_csv("titanic.csv")
dataframe = dataframe.drop(columns=['Name', 'Ticket', 'Cabin', 'PassengerId', 'Pclass', 'Sex','Embarked'])

# Load the datasets
axis_x = 1 # Age
axis_y = 4 # Fare

print(dataframe.head())
X = dataframe.values

model = DBSCAN(eps=0.3, min_samples=25)
model.fit(X)
print(model.labels_)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()

unique_labels = set(model.labels_) 
colors = ['y', 'b', 'g', 'r'] 
core_samples_mask = np.zeros_like(model.labels_, dtype=bool) 
core_samples_mask[model.core_sample_indices_] = True

for k, col in zip(unique_labels, colors): 
    if k == -1: 
        # Black used for noise. 
        col = 'k'
  
    class_member_mask = (model.labels_ == k) 
  
    xy = X[class_member_mask & core_samples_mask] 
    plt.plot(xy[:, axis_x], xy[:, axis_y], 'o', markerfacecolor=col, 
                                      markeredgecolor='k',  
                                      markersize=6) 
  
    xy = X[class_member_mask & ~core_samples_mask] 
    plt.plot(xy[:, axis_x], xy[:, axis_y], 'o', markerfacecolor=col, 
                                      markeredgecolor='k', 
                                      markersize=6) 

plt.title('number of clusters: %d' %len(set(model.labels_))) 
plt.show() 
```
