<h1> Bubble Chart </h1>

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

dframe = pd.read_csv('titanic.csv')
dframe = dframe.drop(columns=['Name','Ticket', 'Cabin'])

list_rows = []

for pclass in (1, 2, 3):
    for gender in ('male', 'female'):
        one_row = {}
        one_row['pclass'] = pclass
        one_row['gender'] = gender

        ba1 = dframe['Pclass']==pclass
        ba2 = dframe['Sex']==gender
        
        one_row['average_age'] = np.round(np.mean(dframe[ba1&ba2]['Age']), 3)
        list_rows.append(one_row)

df = pd.DataFrame.from_dict(list_rows)


sns.relplot(x='pclass', y='gender', size='average_age', sizes=(50, 250), data=df)
plt.show()
```

<h1> Pie Chart </h1>

```python
# Pie chart
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


dframe = pd.read_csv('titanic.csv')
dframe = dframe.drop(columns=['Name','Ticket', 'Cabin'])


# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['First', 'Second', 'Third']
sizes = []

sizes.append(np.sum(dframe['Pclass']==1))
sizes.append(np.sum(dframe['Pclass']==2))
sizes.append(np.sum(dframe['Pclass']==3))

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=0)
plt.show()
```

<h1>Bar charts</h1>

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

dframe = pd.read_csv('titanic.csv')
dframe = dframe.drop(columns=['Name','Ticket', 'Cabin'])

list_rows = []

for pclass in (1, 2, 3):
    for gender in ('male', 'female'):
        one_row = {}
        one_row['pclass'] = pclass
        one_row['gender'] = gender

        ba1 = dframe['Pclass']==pclass
        ba2 = dframe['Sex']==gender
        
        one_row['average_age'] = np.round(np.mean(dframe[ba1&ba2]['Age']), 3)
        list_rows.append(one_row)

df = pd.DataFrame.from_dict(list_rows)

ax = sns.barplot(x='pclass', y='average_age', hue='gender', data=df)

# Define some hatch patterns
hatches = ['-', 'x', '-', 'x', '-','x']

# Loop over the bars
for i,thisbar in enumerate(ax.patches):
    # Set a different hatch for each bar
    thisbar.set_hatch(hatches[i])

plt.show()
```

<h1> Scatter Plot 3D</h1>

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


dframe = pd.read_csv('titanic.csv')
dframe = dframe.drop(columns=['Name','Ticket', 'Cabin'])

dframe.head()
fig = plt.figure()
ax = fig.gca(projection='3d')
ax = sns.scatterplot(x='Pclass', y='Fare', size='Age', data=dframe)
plt.show()
```
